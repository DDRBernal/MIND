# Validación cruzada
# Cross Validation

```python
# Importar las librerías
import pandas as pd
import numpy as np
```

```python
# Importar el dataset vinos
df = pd.read_excel('Vinos.xlsx', index_col = 'num')
```

```python
# Ver los 5 primeros registros
df.head()
```

```python
# Convertir color en categoría
df['color']= df['color'].astype('category')
```

```python
# Remplazar color por el código de su categoría
df['color'] = df['color'].cat.codes
df.head()
```

```python
# Crear función que intercambia el orden de dos columnas (no cambia los títulos)
def swap_columns(df, c1, c2):
    df['temp'] = df[c1]
    df[c1] = df[c2]
    df[c2] = df['temp']
    df.drop(columns=['temp'], inplace=True)
```

```python
# Intercambiar las columnas 'quality' y 'color'
swap_columns(df,'quality','color')
```

```python
# Mostrar los primeros 5 registros
df.head(5)
```

```python
# Cambiar también el nombre de las columnas
lista_col = list(df)
lista_col[-1],lista_col[-2] = lista_col[-2],lista_col[-1]
df.columns = lista_col
```

```python
# Mostrar los primeros 5 registros
df.head(5)
```

## Validación cruzada

```python
# Importar las librerías
from sklearn.svm import SVC #Este el se estimador que usaremos 
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, recall_score
```

```python
# Features del dataframe
X = df[df.columns[:-1]]
# Clases del dataframe
Y = df['quality']
```

```python
# Instanciar el modelo
svm = SVC()
# Crear la validación cruzada. 1) Midiendo la exactitud
# Para el scoring
# https://scikit-learn.org/stable/modules/model_evaluation.html

scores1 = cross_val_score(svm,X,Y,cv=10,scoring= 'accuracy')
```

```python
# Imprimir la lista de exactitud y su media
print("Scores_1 = ",scores1)
print('Media de scores1 = ',scores1.mean())
```

```python
# Crear la validación cruzada. 1) Midiendo la sensibilidad ponderada
scores2 = cross_val_score(svm,X,Y,cv=10,scoring='')
```

```python
# Imprimir la lista de sensibilidad y su media
print("Scores_2 = ",scores2)
print('Media de scores2 = ',scores2.mean())
```

```python
# Crear la validación cruzada. 1) Midiendo la sensibilidad promedio
scores3 = cross_val_score(svm,X,Y,cv=10,scoring='recall_macro')
```
